```python
import sys

packages = {
    "python": sys.version,
    "numpy": None,
    "scipy": None,
    "matplotlib": None,
    "numba": None,
    "pytest": None,
    "dask": None,
    "distributed": None,
}

for name in list(packages.keys()):
    if name == "python":
        continue
    try:
        module = __import__(name)
        packages[name] = getattr(module, "__version__", "unknown")
    except Exception as e:
        print(f"[FAIL] {name}: {e}")
        raise

print("\nEnvironment check ")
for k, v in packages.items():
    print(f"{k:12s}: {v}")

print("\nAll imports successfull")

```
